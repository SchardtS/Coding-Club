```julia
using FileIO
using CairoMakie, Dates
```

```julia
function annotate(img; cols::Int =6, rows::Int = 3, start_time::Time =Time(1, 50, 0)  , time_step::Int = 5)
    im_plot = Figure()
    ax = Axis(im_plot[1,1],aspect = DataAspect(), )
    hidedecorations!(ax)
    image!(ax,rotr90(img))
    size_img = size(img)
    row_size = size_img[1]/rows
    cols_size = size_img[2]/cols
    for i in 1:cols
        for j in 1:rows
            x = round(Int,(i-1)*cols_size + cols_size/2)
            y = round(Int,(j-1)*row_size - row_size/5 +row_size)
            text = string(start_time + Dates.Minute(((i-1)*time_step)))[1:end-3]*"h"
            text!(ax,x, y, text= text, color = :white,font = :bold, fontsize =17)
        end
    end

return im_plot
end
```

```julia
test1 = load("testimage1.png");
test2 = load("testimage2.png");
```

```julia
hans = annotate(test1)
```

```julia
save("testimage1_annotated.png",hans)
```

```julia
dieter = annotate(test2, rows = 4, cols = 5)
```

```julia
save("testimage2_annotated.png",dieter)
```
