```python
import numpy as np
import matplotlib.pyplot as plt
from functions import *
```

### Polynomial function

```python
f = lambda x: x**4 - 10*x**2 + 4
df = lambda x: 4*x**3 - 20*x

x0 = np.array([-3, 3, 1, -1])
x_sol = np.empty(x0.shape)
for i in range(x0.shape[0]):
    x_sol[i] = Newton(f, df, x0[i], 1e-6, 100)
    print('Found roots:', x_sol[i], '\n')
```

```python
x = np.linspace(-4, 4, 100)
plt.plot(x, f(x), lw=2)
plt.axhline(0, color='black', lw=1)
plt.scatter(x_sol, f(x_sol), color='red', s=50, zorder=3)
```

### Exponential function

```python
f = lambda x: np.exp(-x**2) + x/10
df = lambda x: -2*x*np.exp(-x**2) + 1/10

x0 = -1
x_sol = Newton(f, df, x0, 1e-6, 100)
print('Found root:', x_sol, '\n')
```

```python
x = np.linspace(-4, 4, 100)
plt.plot(x, f(x), lw=2)
plt.axhline(0, color='black', lw=2)
plt.scatter(x_sol, f(x_sol), color='red', s=50, zorder=3)
```

### Complex function

```python
f = lambda x: x**5 + 1
df = lambda x: 5*x**4

x0 = np.array([-1, 1+0.5j, 1-0.5j, -0.5+1j, -0.5-1j])
x_sol = np.empty(x0.shape, dtype=complex)
for i in range(len(x0)):
    x_sol[i] = Newton(f, df, x0[i], 1e-6, 100)
    print('Found root:', x_sol[i], '\n')
```

```python
x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)
Z = x[None,:] + 1j*y[:,None]

# x + yj  -- > (x**2 + y**2)**(1/2)

#plt.contourf(x, y, np.abs(f(Z)), 100)
plt.contourf(x, y, np.exp(-np.abs(f(Z))**2), 100)
plt.scatter(x_sol.real, x_sol.imag, color='red')
plt.colorbar()
```

### Two dimensions

```python
f = lambda x: np.array([np.exp(-x[0]**2 - x[1]**2) - 1/2*x[1], x[0]**2 - x[1]])
df = lambda x: np.array([[-2*x[0]*np.exp(-x[0]**2 - x[1]**2), -2*x[1]*np.exp(-x[0]**2 - x[1]**2) - 1/2],
                         [2*x[0], -1]])

#x0 = np.array([[1,1],
#               [-1,1]])
x0 = np.random.uniform(-2, 2, (100,2))
x_sol = np.empty(x0.shape)
sol1 = []
sol2 = []
for i in range(x0.shape[0]):
    x_sol[i,:] = Newton(f, df, x0[i,:], 1e-6, 100)
    print('Found root:', x_sol[i,:], '\n')
    if x_sol[i,0] < 0:
        sol1.append(x0[i,:])
    if x_sol[i,0] >= 0:
        sol2.append(x0[i,:])

sol1 = np.array(sol1)
sol2 = np.array(sol2)
```

```python
x = np.linspace(-2, 2, 1000)
y = np.linspace(-2, 2, 1000)

X, Y = np.meshgrid(x, y)
Z = f(np.array([X, Y]))

plt.figure()

plt.contourf(x, y, np.linalg.norm(Z, axis=0), 100, cmap='coolwarm')
plt.streamplot(X, Y, Z[0], Z[1], density=2, linewidth = np.sqrt(Z[0]**2 + Z[1]**2), color='black')
plt.scatter(x_sol[:,0], x_sol[:,1], color='red', zorder=3)
plt.scatter(sol1[:,0], sol1[:,1], color='yellow', zorder=3)
plt.scatter(sol2[:,0], sol2[:,1], color='green', zorder=3)
```

```python
x = np.linspace(-1.5, 1.5, 30)
y = np.linspace(-0.75, 2.25, 30)

X, Y = np.meshgrid(x, y)
Z = f(np.array([X, Y]))

plt.figure()
plt.contourf(x, y, np.linalg.norm(Z, axis=0), 100, cmap='coolwarm')
plt.quiver(X, Y, Z[0], Z[1], color='black', scale=30)
plt.scatter(x_sol[:,0], x_sol[:,1], color='red', zorder=3)
```
